# Introduction

## National Reference Stations Used for the Data Products

We use ocean temperature data at four long-term national reference stations located around Australia (Fig. 1). These sites are unique in Australia and the Southern Hemisphere in that they have been occupied weekly to monthly since the 1940s and 1950s (Roughan et al., 2022a; Lynch et al., 2014).


The western coast site is close to Rottnest Island (ROT, 32 S) offshore from Perth, Western Australia, in approximately 55 m of water, and temperatures here are influenced by the Leeuwin Current.


The southern site is close to Maria Island (MAI, 42.6 S), Tasmania, in approximately 90 m of water and is situated in the southern extension region of the EAC. An additional two eastern coast sites are located at Port Hacking close to Sydney (PHA and PHB/PH100, 34.1 S), New South Wales, in approximately 50 and 100 m of water, respectively, downstream of the typical EAC separation zone in what are usually eddy-dominated waters.

The ROT, MAI, PHA and PHB/PH100 sites are the longest sub-surface ocean time series in Australian coastal waters and are part of Australia’s national reference station network, a network of long-term sampling sites (presently seven) around Australia (Lynch et al., 2014).

Unlike many long-term ocean temperature datasets, measurements have been collected at multiple depths through the water column at PHA, PHB/PH100, MAI and ROT since the 1940 and 1950s.

The optimal depths (or binned depths) chosen for the data products described here are shown in Fig. 2. These optimal depths were determined using binned data to establish the depths where data availability is highest over the full record (Hemming et al., 2020).

Originally, sampling was boat-based and weekly to monthly. In 2008/2009, the ROT, MAI and PHB/PH100 sites were incorporated into the Integrated Marine Observing System (IMOS) National Reference Station network (Lynch et al., 2014), and continuously recording thermistor moorings have been deployed alongside monthly boat-based sampling.

All moored data have continued to be collected through IMOS since 2008/2009 at a sampling interval of 5–60 min at multiple depths through the water column.

## Defining Extreme temperatures

A marine heatwave (MHW) is defined here as a prolonged discrete anomalously warm event as per the definition of Hobday et al., (2016). If ocean temperatures are warmer than the daily 90th percentiles for 5 days or more they are classed as MHWs. Anything shorter than 5 days is classed as a heat spike (HS).

For marine cold spells (MCSs) and cold spikes (CSs) the same methodology is applied, but instead for temperatures that are lower than the daily 10th percentiles.

Extreme ocean temperature events (MHWs, MCSs, HSs, CSs) have characteristics that can be determined such as duration, onset and decline, and intensity.  

<br>

<img src='https://ohw2024.s3.ap-southeast-2.amazonaws.com/mhw.png' width='600' height='340'>

Source: marineheatwaves.org

<br>

The difference between the daily-varying climatological mean and the 90th or 10th percentiles (depending on event type) is used to categorize extreme ocean temperature events. Multiples of this  range correspond with 4 different categories: 'Moderate', 'Strong', 'Severe', 'Extreme'. The below figure taken from Hobday et al., (2018) illustrates this method of categorising MHWs.

Please note that when I say 'Extreme ocean temperature events', I am referring to MHWs/MCSs/HSs/CSs. Not just 'Extreme' MHW/MCS events as per this definition.

## Downloading files

The below function and code is used to download the AMDOT-EXT files at Maria Island (MAI090) and Port Hacking (PH100)

<img src='https://ohw2024.s3.ap-southeast-2.amazonaws.com/important.png' width='30' height='30'> 

**Note:** you can skip to the next bit to load the files straight from S3. But for this tutorial, I chose to download the files to my directory as loading straight from S3 is slower.

**Tip:** In AWS, folders are called 'buckets' and files are called 'objects'.

```python
# Function to download files from S3
def download_files_from_s3(bucket_name, s3_folder, local_dir):
    # Create a boto3 session
    anonymous_session = boto3.Session()
    # Create a client with anonymous access
    s3 = anonymous_session.client(
        's3',
        config=Config(signature_version=UNSIGNED)
    )
    
    # List objects within a given S3 bucket
    objects = s3.list_objects_v2(Bucket=bucket_name, Prefix=s3_folder)
    
    if 'Contents' in objects:
        for obj in objects['Contents']:
            file_name = obj['Key'].split('/')[-1]
            local_file_path = os.path.join(local_dir, file_name)
            
            # Download the file
            s3.download_file(bucket_name, obj['Key'], local_file_path)
            
            print(f'Downloaded {file_name} to {local_file_path}')

    else:
        print("No files found in the specified S3 folder.")
```

# Extracting data and Saving

The below code does the following:

* Determines the longest MHW at the near-surface
* Selects the temperature and 90th percentiles during this MHW
* saves this output into a .CSV

```python
# Select surface MHWs
SurfMHWs = MAI090['TEMP_EXTREME_INDEX'][:, 0].values == 12

# Select surface MHW duration
SurfDuration = MAI090['MHW_EVENT_DURATION'][:, 0].values

# Determine longest surface MHW
LongestDuration = np.nanmax(SurfDuration[SurfMHWs])

# Get boolean to select longest surface MHW
LongestSelection = SurfDuration == LongestDuration
```

```python
LongestDuration
```

```python
# Select times during the longest surface MHW
LongestTimes = MAI090['TIME'][LongestSelection]

# Select temperature and the 90th percentiles during marine heatwave
data_selection = MAI090.sel(TIME=slice(np.min(LongestTimes.values), np.max(LongestTimes.values)))
```

```python
data_selection
```

```python
# Temperature
TEMP_selection = data_selection.TEMP.to_pandas()
# Rename columns
TEMP_selection.rename(columns={2.0: 'TEMP 2m'}, inplace=True)
TEMP_selection.rename(columns={21.0: 'TEMP 21m'}, inplace=True)
```

```python
TEMP_selection
```

```python
# 90th percentiles
PER90_selection = data_selection.TEMP_PER90.to_pandas()
# Rename columns
PER90_selection.rename(columns={2.0: 'PER90 2m'}, inplace=True)
PER90_selection.rename(columns={21.0: 'PER90 21m'}, inplace=True)
```

```python
PER90_selection
```

```python
# Merge data frames for exporting as spreadsheet
data2save = pd.merge(TEMP_selection, PER90_selection, left_on='TIME', right_on='TIME')
```

The steps to link your Github repository with Zenodo can be found here: https://docs.github.com/en/repositories/archiving-a-github-repository/referencing-and-citing-content

Consider whether it's relevant to merge code from multiple repositories into one repository for publishing. This will be more efficient as it limits the amount of DOIs that you will need to mint, and you may get more usage and citations.

## Create a Citation.cff file

Once you hjave organised your repository, you can add a citation.cff file. Including a citation.cff file in a GitHub repository prior to releasing and creating a DOI using Zenodo offers several benefits:

* Standardized Citation: The citation.cff file follows the Citation File Format (CFF) standard, providing a consistent and machine-readable format for describing the software project or research output. It ensures that the citation information is well-structured and easily understood by both humans and automated systems.

* Easy DOI Creation: Zenodo, a popular DOI service integrated with GitHub, can automatically generate a DOI for a repository based on the information provided in the citation.cff file. By including this file, you streamline the process of DOI creation, making it more convenient and time-efficient.

* Proper Attribution: The citation.cff file includes essential details such as authors, title, description, version, and license information. By providing this information, you ensure that proper attribution is given to the creators of the repository and that others can easily cite and reference your work.

* Improved Discoverability: Including a citation.cff file enhances the discoverability of your repository and associated research output. The standardized format and comprehensive metadata provided in the file make it easier for search engines, indexing services, and other platforms to accurately capture and represent your work.

Also, if your repository is hosted on GitHub, they will show the citation information in the sidebar which makes it easy for visitors to cite your software or dataset correctly.

More information is available here: https://github.com/citation-file-format/citation-file-format

## Create a DOI

The easiest way to create a DOI is to use Zenodo and logging in with your github account. When you are happy with the state of your linked repository, proceed with the steps described here: https://docs.github.com/en/repositories/archiving-a-github-repository/referencing-and-citing-content

Everytime you create a new release in the Github repository, a new DOI will be isued.

Note: Your repository must be public at this step.

## Other options

* Dryad - https://datadryad.org/stash. Example: https://datadryad.org/stash/dataset/doi:10.5061%2Fdryad.jwstqjqdh
* Figshare (for profit) - https://figshare.com/. Example: https://figshare.com/articles/software/Data_Product_tutorials_for_Roughan_et_al_Multi-decadal_ocean_temperature_time-series_and_climatologies_from_Australia_s_long-term_National_Reference_Stations_2022_/18232691

Dryad allows for 300GB total repository limit for files (individual files should be < 10GB), Zenodo allows 50GB.
